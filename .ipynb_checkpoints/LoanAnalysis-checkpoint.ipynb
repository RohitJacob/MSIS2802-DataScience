## Calculating the Percentage Change

In the range 2000 to 2016, some countries don't have any financial records. So to ensure we don't rule out their presence by dropping NaN elements, the dataframe will search for the country's first financial record and we save it in a new Dataframe for comparison

```python
newdf = pd.DataFrame(columns=['Country', 'EarliestRecord', '2016Jan'])
for column in df:
    newdf.loc[len(newdf)]=[column, df[column][df[column].last_valid_index()], df[column]['2016Jan']]
```

```python
newdf['ChangeInPercent'] = newdf['2016Jan']-newdf['EarliestRecord']
```
