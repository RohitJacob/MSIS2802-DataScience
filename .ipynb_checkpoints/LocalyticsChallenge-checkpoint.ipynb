### Building A Variable Selection Model
<p>We've Got A Good Idea of What The Data is About. Now, Using a Little Machine Learning, I wanted to Learn What Are The Main Variables That Affect My Target: <b>Amount Spent</b>
</p>
<p>I chose <b>Lasso Regression</b> as my weapon of choice because it is a commonly used variable selection tool.

#### The Following Functions convert all my variables in my Prediction Dataframe to a integer code. This in turn is then preprocessed using sklearn's preprocessing tool that gets our variables ready to feed into the LassoLarsCV model.

```python
ml = df.dropna() 
predictiondf= ml[['age', 'category', 'gender', 'location.city', 'location.state', 'marital_status']]
predictiondf
target = ml.amount
```

```python
def agegroup(agerange):
    if(agerange == ageCount.index[0]):
        return 0 
    elif(agerange == ageCount.index[4]):
        return 1
    elif(agerange == ageCount.index[1]):
        return 2
    elif(agerange == ageCount.index[2]):
        return 3
    elif(agerange == ageCount.index[3]):
        return 4
```

```python
categories = 
```

```python
def catgroup(cat):
    if(cat == categories.index[0]):
        return 0 
    elif(cat == categories.index[1]):
        return 1
    elif(cat == categories.index[2]):
        return 2
    elif(cat == categories.index[3]):
        return 3
    elif(cat == categories.index[4]):
        return 4
```

```python
def genderclass(g):
    if(g == 'M'):
        return 1 
    elif(g == 'F'):
        return 2
    elif(g == 'U'):
        return 0
```

```python
def maritalclass(m):
    if(m == 'married'):
        return 1 
    elif(m == 'single'):
        return 0
```

```python
uniqueStates = pd.DataFrame(predictiondf['location.state'].unique())
def stateclass(s):
    for i in range(0, uniqueStates.shape[0]):
        if s == uniqueStates[0][i]:
            return i
```

```python
def cityclass(s):
    for i in range(0, cities.shape[0]):
        if s == cities.index[i]:
            return i
```

```python
predictiondf.age = predictiondf.age.apply(lambda x: agegroup(x))
predictiondf.category = predictiondf.category.apply(lambda x: catgroup(x))
predictiondf.gender = predictiondf.gender.apply(lambda x: genderclass(x))
predictiondf.marital_status = predictiondf.marital_status.apply(lambda x: maritalclass(x))
predictiondf['location.state'] = predictiondf['location.state'].apply(lambda x: stateclass(x))
predictiondf['location.city'] = predictiondf['location.city'].apply(lambda x: cityclass(x))
```

```python
for col in predictiondf.columns:
    predictiondf[col]=preprocessing.scale(predictiondf[col].astype('float64'))
```

```python
predictiondf.head()
```

#### Now We Can Plug In Variables Into The Model 

```python
# split data into train and test sets
pred_train, pred_test, tar_train, tar_test = train_test_split(predictiondf, target,test_size=0.7, random_state=123)
# specify the lasso regression model
model=LassoLarsCV(cv=10, precompute=False).fit(pred_train,tar_train)
dict(zip(predictiondf.columns, model.coef_))
```

```python
m_log_alphas = -np.log10(model.alphas_)
ax = plt.gca()
plt.plot(m_log_alphas, model.coef_path_.T)
plt.axvline(-np.log10(model.alpha_), linestyle='--', color='k',
            label='alpha CV')
plt.ylabel('Regression Coefficients')
plt.xlabel('-log(alpha)')
plt.title('Regression Coefficients Progression for Lasso Paths')
```

### Outcome: Looks like Marital Status Has An Effect on the Target.
I've ignored the sign of the coefficient only because the values were coded into 1s and 0s and may affect the polarity of the model.

### Let's See How They Spend...

```python
devicefunding = df[df['event_name'] == 'Fund Project'].groupby(['device'])['amount'].sum().\
                                       to_frame().sort_values(by='amount', ascending=False)
```

```python
devices['index'] = devices.index
devicefunding['index']=devicefunding.index
deviceAvg = pd.merge(devices, devicefunding, on='index', how='left')
deviceAvg['Average']= deviceAvg['amount']/deviceAvg['device']
deviceAvg.index = deviceAvg['index']
# deviceAvg['Number Of devices'] = deviceAvg['device']
deviceAvg = deviceAvg[['amount','Average']]
deviceAvg['amount'].plot(kind='bar', color=('whitesmoke','lightsage'))
```
