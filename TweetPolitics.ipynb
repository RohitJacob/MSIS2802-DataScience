```python
def cleaner(x):
    ml= []
    y = x.split(" ")
    for i in range(len(y)-1):
        y[i] = re.sub('()[!#?:\'\"\\.,]', '', y[i])
        ml.append(y[i])
        if 'RT' in y[i] or '@' in y[i] or 'http' in y[i] or '&amp;' in y[i] or y[i].lower() == "hillary" or y[i].lower() == "hillarys" or y[i].lower() == "trump" or y[i].lower() == "trumps" or y[i].lower() == "clinton" or y[i].lower() == "clintons" or y[i].lower() == "donald" or y[i].lower() == "donalds":
            ml.remove(y[i])
    s = ''.join((str(e)+" ") for e in ml)
    return s
```

```python
hilTweets['tweets'] = hilTweets[0]
hilTweets = hilTweets.drop(hilTweets.columns[0], axis=1)
hilTweets.tweets = hilTweets.tweets.apply(cleaner)
```

```python
trumpTweets['tweets'] = trumpTweets[0]
trumpTweets = trumpTweets.drop(trumpTweets.columns[0], axis=1)
trumpTweets.tweets = trumpTweets.tweets.apply(cleaner)
```
